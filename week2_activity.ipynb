```python
plt.figure(figsize=(10,5))
plt.scatter(df['Rank'],df['Happiness Score'])
plt.xlabel('Rank') # Add x-axis label
plt.ylabel('Happiness Score') # Add y-axis label
plt.title('Happiness Score vs Rank') # Add a title
plt.show() # Show the plot
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Rank'],df['Happiness Score'])
plt.xlabel('Rank')
plt.ylabel('Happiness Score')
plt.title('Happiness Score vs Rank')
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Rank'],df['Happiness Score'],color='purple',marker='*')
plt.xlabel('Rank')
plt.ylabel('Happiness Score')
plt.title('Happiness Score vs Rank')
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Rank'],df['Happiness Score'],color='yellow',marker='*')
plt.xlabel('Rank')
plt.ylabel('Happiness Score')
plt.title('Happiness Score vs Rank')
plt.grid(True)
```
